# Analysis of Brandeis Courses 
Today we will analyze a data set of Brandeis courses.
This covers the period Fall 2004 to Summer 2020



````python
import csv
from matplotlib import pyplot as plt
import numpy as np
````

## Reading in an excel file with csv
the csv package allows one to read an excel file (stored as a comma-separated-values csv file)
as a list of dictionaries where the keys are the column headings

````python
cfile = open('courses.csv','r',encoding='utf-8')
courses = list(csv.DictReader(cfile))
print(len(courses),'courses were read')
courses[144]
````

## Cleaning the data
The numerical fields, enr, num, code are stored as strings.
It is better to convert them to integers. 
The following code loops through each dictionary and changes the enr and num fields to be integers

````python
for course in courses:
    course['enr'] = int(course['enr'])
    course['num'] = int(course['num'])
courses[144]
````

````python
plt.figure(figsize=(20,20))
nums = [course['num'] for course in courses]
enrs = [course['enr'] for course in courses]
plt.scatter(nums,enrs,s=enrs)
````

## Python sets
We can use the Python set comprehension to find all of the semesters represented.
The set comprehension has the form:
``` python
{ EXPR(VAL) for VAL in LIST}
```
The following code shows how we find all of the code/sem tuples with no duplicates

````python
semesters = {(course['code'],course['sem']) for course in courses}
semesters
````

````python
x = courses[0]
for x in courses:
    x['id'] = x['subject']+" "+str(x['num'])+x['suf']+"-"+x['section']
x
````
