# Data Analysis in Python
with matplotlib and csv

First we import some library that will allow us to read in an excel file (stored in csv or comma-separated values format) as a list of dictionaries.

```python
import csv
from matplotlib import pyplot as plt
import numpy as np
```

## using DictReader
This is the standard technique for reading an excel file in as a list of dictionaries.

```python
with open('baby100.csv','r',encoding='utf-8') as bfile:
    data = list(csv.DictReader(bfile))
print(len(data),'names were read')
```

```python
years = list(range(1880,2020))
print(years)
```

```python
def name_counts(name,sex):
    return([(int(d['year']),float(d['percent'])*1000) for d in data if d['name']==name and d['sex']==sex])
```

```python
names = name_counts('Gina','F')
names
```

```python
def pick(vals,k):
    return [val[k] for val in vals]
```

```python
print(pick(names,0))
```

```python
print(pick(names,1))
```

```python
plt.plot(pick(names,0),pick(names,1))
plt.grid()
plt.title("popularity of Gina")
```

```python
def plot_name(name,sex):
    vals = name_counts(name,sex)
    plt.plot(pick(vals,0),pick(vals,1),label=name+sex)
```

```python

plot_name('James','M')
plot_name('William','M')
plot_name('Timothy','M')
plot_name('Mary','F')
plot_name('Emma','F')
plot_name('Gloria','F')
plt.legend()
plt.grid()
```

```python
girls1880 = "Mary Anna Emma Elizabeth Minnie".split()
for name in girls1880:
    plot_name(name,'F')
plt.legend()
plt.grid()
```
