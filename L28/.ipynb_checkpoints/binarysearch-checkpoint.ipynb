# Binary Search

```python
import random
```

```python
def makelist(n):
    return [random.randint(1,10000) for i in range(n)]
```

```python
vals = sorted(makelist(100))
print(vals)
```

```python
vals[99]
```

```python
vals = sorted(makelist(1000))
len(vals)
```

```python
vals[46]
```
