Here is the code to plot a sine function


````python
from matplotlib import pyplot as plt
import numpy as np
````

````python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.grid()
plt.title("Sine function",fontsize=32)
plt.xlabel("radians")
plt.ylabel("function value")
````

````python
x = np.arange(-10, 10, 0.01)
y = np.sin(np.cos(x))-np.cos(np.sin(x))
plt.plot(x, y)
plt.grid()
plt.title("Sin(cos)-cos(sin) function",fontsize=32)
plt.xlabel("radians")
plt.ylabel("function value")
````
